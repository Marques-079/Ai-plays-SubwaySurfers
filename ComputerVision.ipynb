```python
import cv2
import os

def fast_extract_frames(video_path, output_dir, start_trim=0, end_trim=0, interval_sec=2.4):
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration_sec = total_frames / fps

    start_frame = int(start_trim * fps)
    end_frame = int((duration_sec - end_trim) * fps)
    interval_frames = int(interval_sec * fps)

    print(f"âš¡ Accelerated frame extraction starting...")
    print(f"ğŸ¥ Total duration: {duration_sec:.2f}s | FPS: {fps:.2f}")
    print(f"ğŸ“¸ Saving every {interval_sec}s from {start_trim}s to {duration_sec - end_trim}s")

    count = 0
    for frame_number in range(start_frame, end_frame, interval_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        ret, frame = cap.read()
        if not ret:
            break

        filename = os.path.join(output_dir, f"frame_{count:04d}.jpg")
        cv2.imwrite(filename, frame)
        print(f"ğŸ–¼ï¸ Saved {filename}")
        count += 1

    cap.release()
    print(f"âœ… Done: {count} screenshots saved to '{output_dir}'")


# === USER CONFIGURABLE ===
if __name__ == "__main__":
    VIDEO_PATH = "/Users/marcus/downloads/video3.mp4"
    OUTPUT_DIR   = "train_screenshots3"      # Folder to save cropped frames
    TRIM_START   = 15                     # Seconds to trim from the beginning
    TRIM_END     = 39                       # Seconds to trim from the end
    INTERVAL_SEC = 2                  # Time between frames to save

    fast_extract_frames(
        VIDEO_PATH,
        OUTPUT_DIR,
        start_trim=TRIM_START,
        end_trim=TRIM_END,
        interval_sec=INTERVAL_SEC
    )
```
